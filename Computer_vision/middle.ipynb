```python
# 비디오 데이터 읽어내는 코드
import cv2
import sys

cap = cv2.VideoCapture('img/video.mp4')

if not cap.isOpened():
    sys.exit('camera not found')

# 제대로 프레임 읽으면 ret이 True 아니면 False
# 'q'를 누를 때까지 반복함
while True:
    ret, frame = cap.read()
    if not ret:
        print('cannot read capture img')
        break
    cv2.imshow('show img', frame)
    key = cv2.waitKey(1)
    if key == ord('q'):
        break

cap.release()   # 오픈한 캡쳐 객체 해제
cv2.destroyAllWindows()
```

```python
#  그래픽 기능: 영상에 도형을 그리고 글씨 쓰기

import cv2
import sys
import os


img = cv2.imread('img/bird.jpg')
if img is None:
    sys.exit('file not exist')

cv2.rectangle(img, (50,20), (450,400), (0,0,255), 3)
cv2.putText(img, 'bird', (250,50), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,0,0),2)

cv2.imshow('draw', img)
cv2.waitKey()

cv2.destroyAllWindows()
```

```python
# 마우스를 클릭하여 사각형 그리기

import cv2
import sys
import os


img = cv2.imread('img/bird.jpg')
if img is None:
    sys.exit('file not exist')

def draw(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.rectangle(img,(x,y), (x+200,y+200), (0,0,255),3)
    elif event == cv2.EVENT_RBUTTONDOWN:
        cv2.rectangle(img, (x,y), (x+100, y+100), (255,0,0),1)
    cv2.imshow('Drawing', img)


cv2.namedWindow('Drawing')
cv2.imshow('Drawing', img)

cv2.setMouseCallback('Drawing', draw)

while(True):
    if cv2.waitKey(1) == ord('q'):
        cv2.destroyAllWindows()
        break
```

```python
# 마우스 드래그로 도형 크기 조절하기

import cv2
import sys
import os


img = cv2.imread('img/bird.jpg')
if img is None:
    sys.exit('file not exist')

def draw(event, x, y, flags, param): 
    global a, b
    if event == cv2.EVENT_LBUTTONDOWN:      # 마우스를 눌렀을 때 좌표
        a, b = x, y
    elif event == cv2.EVENT_LBUTTONUP:      # 마우스를 땠을 때
        cv2.rectangle(param, (a, b), (x, y), (255,0,0),3)       # 사각형을 그림
    cv2.imshow('Drawing', img)


cv2.namedWindow('Drawing')
cv2.imshow('Drawing', img)
cv2.setMouseCallback('Drawing', draw, img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# [실습 5]페인팅 기능 만들기

import cv2
import sys
import os

img = cv2.imread('img/bird.jpg')

if img is None:
    sys.exit('file not found')


BrushSize = 5
LColor, RColor = (255,0,0), (0,0,255)   # Blue, Red (B, R, G)

def painting(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img, (x,y), BrushSize, LColor, -1)
    elif event == cv2.EVENT_RBUTTONDOWN:
        cv2.circle(img, (x,y), BrushSize, RColor, -1)
    elif event == cv2.EVENT_MOUSEMOVE and flags == cv2.EVENT_FLAG_LBUTTON:
        cv2.circle(img, (x,y), BrushSize, LColor, -1)
    elif event == cv2.EVENT_MOUSEMOVE and flags == cv2.EVENT_FLAG_RBUTTON:
        cv2.circle(img, (x,y), BrushSize, RColor, -1)
    cv2.imshow('Painting', img)


cv2.namedWindow('Painting')
cv2.imshow('Painting', img)

cv2.setMouseCallback('Painting', painting)

while(True):
    if cv2.waitKey(1) == ord('q'):
        cv2.destroyAllWindows()
        break
```

```python
# [실습 7] 글씨 영역 찾기
import cv2
import os

img = cv2.imread('img/handwritten.jpg')

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = cv2.GaussianBlur(gray, (5,5), 0)

_, bin_img = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

mser = cv2.MSER_create()
regions, _ = mser.detectRegions(bin_img)

for bbox in regions:
    x,y,w,h = cv2.boundingRect(bbox)
    cv2.rectangle(img, (x,y), (x+w, y+h), (0,255,0), 2)

cv2.imshow('img', img)
cv2.waitKey()
cv2.destroyAllWindows()
```
